### Question 1

A continuous time sinusoidal (sine) signal x(t) has a peak-to-peak value of 6, frequency of 16 Hz and a DC offset value of -2.

Generate and Plot 32 samples of this signal using Python if it is sampled with fs = 128 Hz.

X-axis unit should be in "seconds" and labeled.
Grid should be on
​Save your figure as "png" file and upload via button below.

Grading:

- Correct x-axis unit and label (6)

### Solution

- $ f_{real} = 16 $ Hz
- Sampling frequency = $ f_s = 128 $



```python
%pylab inline
#from numpy import *
#from matplotlib import *
#from matplotlib.pyplot import *

fs = 128.
fr = 16
n = arange(32)
t = n /fs
x = 3*sin(2*pi*fr*t)-2

plot(t,x)
xlabel("Time (sec)")
grid(True)
```

### Question 2

For unknown number of FFT points M, it is known that  eleventh (k=10) basis function has a real frequency of 50 Hz for a sampling frequency of 5 kHz. 

Generate and plot the real part of 21st (k=20) basis function for a length of (almost) 20 ms.

- X-axis must be in time (seconds) units. Figure should involve all important information.

- ​Save your figure as "png" file and upload via button below.

Grading:

- Correct Waveform Plotted (13)

- Correct x-axis unit and label (4)

- Grid, Title (3)


### Solution

By using the first sentence, we can calculate M:

$$ \Omega_k = 2\pi \frac{k}{M} = 2 \pi \frac{f_{real}} {f_s} $$
$$ \frac{k}{M} = \frac{f_{real}} {f_s} $$

We are given $k=10$, $f_r = 50 $ Hz and $f_s = 5 $ kHz. So:
$$ \frac{10}{M} = \frac{50} {5000} \rightarrow M = 1000 $$ 

So the number of FFT points is 1000. Then, the 21st basis function is
$$ u_{20}(n) = e^{j*2*pi*20*n/1000} $$
and real part is directly the cosine component.

```python
fs = 5000.
n = arange(0.02*fs) # 20 ms
t = n/fs
M = 1000
u20real = cos(2*pi*20*n/M)

plot(t,u20real)
xlabel("Time (sec)")
grid(True)
title("Real Component of $u_{20}$ (n)")

```

$$ f(t) = at + b $$

$$ \frac{\Delta f} {\Delta t } =\frac{300 - 100} { 2.5 - 0.5} = \frac{200}{2} = 100 \ Hz/sec$$

So the initial point (t=0) is 50 Hz.

$$ f(0) = 100* 0 + b =50 $$
$$ b=50 $$

$$ f(t) = 100 t + 50 $$

### Question 5

Using the phase function, generate (almost) 10 seconds of x(t) for fs = 4 kHz.

Plot time vs. x(t)  and show ONLY the time interval from 0 to 40 ms.

Save your figure as "png" file and upload it via button below.

Grading:

- Correct Waveform plotted (9)

- Correct x-axis unit, label and limits (6)

```python
# Solution
fs = 4000.
t = arange(10*fs)/fs
phi = 100*pi*(t**2 + t)
x = sin(phi)

plot(t,x)
xlim([0, 0.04])
xlabel("Time (sec)")

```
