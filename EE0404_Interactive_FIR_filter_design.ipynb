```python

#%pylab inline
from matplotlib import *
from scipy.signal import *
from IPython.html.widgets import *

        
from matplotlib.pyplot import *
from numpy import *
%matplotlib inline
rcParams['figure.figsize']=(16,4)
```

```python
def sd(order=10,cutoff=0.5):
    print "Filter order: ", order
    print "Filter cut-off: ", cutoff, 'pi'
    b = firwin(order+1,cutoff)
    w,H = freqz(b,1)
    subplot(1,2,1)
    plot(w/pi,abs(H))
    hold(True)
    plot([cutoff,cutoff],[0, 1.2],'--r')
    grid(True)
    ylabel('|H($e^{jw}$)|')
    xlabel('Normalized Freq. $\Omega/\pi$')
    title('Freqeuency Response')
    
    subplot(1,2,2)
    n = arange(len(b))
    stem(n,b)
    title('Impulse Response')
```

```python
#interaction.interact_manual(sd,order=(1,50,5),cutoff=(0.1,1,.05))
interact(sd,order=(1,50,5),cutoff=(0.1,1,.05))
```
