# Discrete Fourier Transform
- Euler Formula
- Basis functions
- Real frequencies of basis functions
- Frequency resolution


```python
import numpy as np
import matplotlib.pyplot as plt

def pol2cart(r,theta):
    return r*np.cos(theta), r*np.sin(theta)

thetas = np.linspace(0,2*np.pi,100)
[xx, yy]= pol2cart(1,thetas)
```

```python
from IPython.display import display, Math, Latex
display(Math(r'\sum\limits_{n=0}^{M-1} x(n)e^{j \frac{2\pi}{M}nk},\ \ \ k=\{0,1,...,M-1\}'))
```

```python
%matplotlib inline
P = np.exp(1j*.75*np.pi)
arc_theta = np.linspace(0,.75*np.pi,20)
[arc_xx, arc_yy]= pol2cart(.3,arc_theta)

plt.plot(xx,yy,'--');
plt.axes().set_aspect('equal')
plt.axis([-1.5 ,1.5, -1.5, 1.5 ])
plt.grid(True)
plt.hold(True)
plt.plot(arc_xx,arc_yy);
plt.plot(np.real(P),np.imag(P),'or');
plt.plot([0,np.real(P)],[0, np.imag(P)],'--b');
```

# Question
Generate and plot cosine and sine functions that are sampled at fs=60 Hz and have a real frequency 3 Hz.

- Are there any missing parameters? 
- Duration? Let’s pick it as 1 second.



```python
fs = 60
n = np.arange(0,60)
fr = 3
euler = np.exp(1j*2*np.pi*fr*n/fs)
plt.plot(n,np.real(euler));
plt.grid(True)
plt.hold(True)
plt.plot(n,np.imag(euler));
plt.title('Real and Complex Parts of Euler');
plt.xlabel('samples - (n)');
plt.ylabel('Magnitude');
plt.ylim([-2,2]);
```

# Oh dear, DFT!


```python
display(Math(r'y(k) = \sum\limits_{n=0}^{M-1} x(n)\left(e^{j \frac{2\pi}{M}nk}\right)^*,\ \ \ k=\{0,1,...,M-1\}'))
display(Math(r'y(k) = \sum\limits_{n=0}^{M-1} x(n)e^{-j \frac{2\pi}{M}nk},\ \ \ k=\{0,1,...,M-1\}'))
display(Math(r'y(k) = \sum\limits_{n=0}^{M-1} x(n)\left[(\cos\left(e^{j \frac{2\pi}{M}nk}\right)-j\sin\left(e^{j \frac{2\pi}{M}nk}\right)\right]'))
display(Math(r'y(k) = \sum\limits_{n=0}^{M-1} x(n)\cos\left(e^{j \frac{2\pi}{M}nk}\right) - j\sum\limits_{n=0}^{M-1} x(n)\sin\left(e^{j \frac{2\pi}{M}nk}\right)'))
```

- M is the number of observed values of signal x(n)
 - n = 0, 1, …, M-1
- So here, we have M points of DFT calculation.
 - k = 0, 1, …, M-1
